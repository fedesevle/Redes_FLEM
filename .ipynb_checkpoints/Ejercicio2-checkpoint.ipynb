```python
# Cargamos los archivos de datos sobre los delfines, tanto la red como la informacin de gnero de los delfines.

import matplotlib.pylab as plt
import networkx as nx
import numpy as np
import pandas as pd

G = nx.read_gml('dolphins.gml')

def ldata(archive):
    f=open(archive)
    data=[]
    for line in f:
        line=line.strip()
        col=line.split()
        data.append(col)
    return data

a = ldata('dolphinsGender.txt')

# Ahora vamos a incorporar el dato de gnero al objeto que contiene la red

for dolph in a:
    G.nodes[dolph[0]]['Gender'] = dolph[1]
    
# Luego tenemos que graficar las redes para diferentes layouts. De estos me parece que el ms apropiado es el cuarto, porque
# muestra mejor la segregacin de los delfines por gnero.

options = {
     'node_color':["blue" if Gender=="f" else "red" for Gender in nx.get_node_attributes(G, "Gender").values()],
     'node_size': 30,
     'width': 2,
}
plt.subplot(231)
nx.draw_random(G, **options)
plt.subplot(232)
nx.draw_circular(G, **options)
plt.subplot(233)
nx.draw_spectral(G, **options)
plt.subplot(234)
nx.draw_spring(G, **options)
plt.subplot(235)
nx.draw_kamada_kawai(G, **options)
plt.show()

# Ac estamos sacando la cuenta de cuntos delfines de cada gnero existen. La diferencia de 4 delfines entre el total y
# la suma de los generos se debe a los cuatro delfines que no tienen un gnero asignado.

counter_m = 0
counter_f = 0

for line in a:
    if line[1] == 'm':
        counter_m = counter_m +1
    else: 
        if line[1] == 'f':
            counter_f = counter_f +1

print('cantidad de delfines machos: '+str(counter_m))
print('cantidad de delfines hembras: '+str(counter_f))
print('cantidad total de delfines: '+str(len(a)))

# A continuacin vamos a determinar cuntos enlaces de tipo inter-gnero existen entre los delfines de la red

contad_diff = 0
c = dict(G.edges)

for line in c:
    if G.nodes[line[1]] != G.nodes[line[0]]:
        G.edges[line[0],line[1]]['T_enl'] = 'diff'
        contad_diff= contad_diff +1
    else:
        G.edges[line[0],line[1]]['T_enl'] = 'eq'
        
print('Cantidad de enlaces diferentes: '+str(contad_diff))
print('Cantidad total de enlaces: '+str(len(c)))
```

```python
# Vamos a definir una funcin que asigna el gnero al azar y luego calcula la cantidad de enlaces intergnero que quedan 
# formados. El resultado que devuelve es la cantidad de enlaces.

import random

def GeneroAzar(a):
    b = a
    for num in b:
        num[1]='f'
    indices_m = random.sample(range(62), 34)
    for num in indices_m:
        b[num][1] = 'm'
       
    for dolph in b:
        G.nodes[dolph[0]]['Gender'] = dolph[1]
    
    contad_diff = 0
    c = dict(G.edges)
    for line in c:
        if G.nodes[line[1]] == G.nodes[line[0]]:
            G.edges[line[0],line[1]]['T_enl'] = 'diff'
            contad_diff= contad_diff +1
        else:
            G.edges[line[0],line[1]]['T_enl'] = 'eq'
    return contad_diff

# Crearemos una lista de 36 ndices a los cules se les asignar el gnero 'masculino'. Esos nmeros se elegirn al azar de
# entre el total de delfines.
```

```python
# Hacemos las mil iteraciones para crear las redes con distribucin azarosa de enlaces intergnero

Dist_interg = []
for x in range(0,10000):
    dis = GeneroAzar(a)
    Dist_interg.append(dis)
    

#print(Dist_interg)
```

```python
plt.hist(Dist_interg, bins='auto')  # arguments are passed to np.histogram
plt.title("Histogram with 'auto' bins")
plt.show()
```

```python
print('media de la distribucin: '+str(np.mean(Dist_interg)))
print('desviacin standard de la distribucin: '+str(np.std(Dist_interg)))
```
