### Ítem b: Análisis de homofilia

El análisis de la presencia de homofilia se realiza considerando la distribución nula y comparándola con la distribución real de enlaces entre delfines de distinto género.

#### i. Generación de la distribución nula

Para generar una distribución nula se tomó la red original y se reasignaron al azar los géneros de cada uno de los delfines, pero cuidando que la proporción total de machos y hembras se mantuviera constante.

```python
# Acá estamos sacando la cuenta de cuántos delfines de cada género existen. La diferencia de 4 delfines entre el total y
# la suma de los generos se debe a los cuatro delfines que no tienen un género asignado.

counter_m = 0
counter_f = 0

for line in a:
    if line[1] == 'm':
        counter_m = counter_m +1
    else: 
        if line[1] == 'f':
            counter_f = counter_f +1

print('cantidad de delfines machos: '+str(counter_m))
print('cantidad de delfines hembras: '+str(counter_f))
print('cantidad total de delfines: '+str(len(a)))

# A continuación vamos a determinar cuántos enlaces de tipo inter-género (macho con hembra) existen entre los delfines 
#de la red

contad_diff = 0
c = dict(G.edges)

for line in c:
    if G.nodes[line[1]] != G.nodes[line[0]]:
        G.edges[line[0],line[1]]['T_enl'] = 'diff'
        contad_diff= contad_diff +1
    else:
        G.edges[line[0],line[1]]['T_enl'] = 'eq'
        
print('Cantidad de enlaces diferentes: '+str(contad_diff))
print('Cantidad total de enlaces: '+str(len(c)))
```

A efectos de poder generar redes aleatorias, teníamos primero que conocer las cantidades de machos y de hembras. De paso, hemos procedido ya a calcular la cantidad de enlaces y la cantidad de enlaces diferentes para la red real. Esto será importante para el próximo ítem. A continuación procedemos a crear 10000 redes al azar, manteniendo la topología y cantidad total de machos y hembras, pero alterando al azar el género de cada delfín individual.

```python
# Vamos a definir una función que asigna el género al azar y luego calcula la cantidad de enlaces intergénero que quedan 
# formados. El resultado que devuelve es la cantidad de enlaces.

import random

def GeneroAzar(a):
    b = a
    for num in b:
        num[1]='f'
    indices_m = random.sample(range(62), 34)
    for num in indices_m:
        b[num][1] = 'm'
       
    for dolph in b:
        G.nodes[dolph[0]]['Gender'] = dolph[1]
    
    contad_diff = 0
    c = dict(G.edges)
    for line in c:
        if G.nodes[line[1]] != G.nodes[line[0]]:
            G.edges[line[0],line[1]]['T_enl'] = 'diff'
            contad_diff= contad_diff +1
        else:
            G.edges[line[0],line[1]]['T_enl'] = 'eq'
    return contad_diff

# Crearemos una lista de 36 índices a los cuáles se les asignará el género 'masculino'. Esos números se elegirán al azar de
# entre el total de delfines.
```

```python
# Hacemos las mil iteraciones para crear las redes con distribución azarosa de enlaces intergénero

Dist_interg = []
for x in range(0,10000):
    dis = GeneroAzar(a)
    Dist_interg.append(dis)
    

#print(Dist_interg)
```

A partir de los valores de cantidad de enlaces distintos para cada una de las 10000 redes aleatorias se realizó el histograma que se muestra a continuación

```python
plt.hist(Dist_interg, bins='auto')  # arguments are passed to np.histogram
plt.title("Distribución de enlaces inter-género")
plt.show()
```

#### ii. Estimación y comparación

A partir de las redes del gráfico anterior se calcularon su media y su desviación típica. La media se calculó con el valor real de la cantidad de enlaces. Esta comparación muestra que el valor real es muy diferente a la media de la distribución.

```python
print('media de la distribución: '+str(np.mean(Dist_interg)))
print('desviación standard de la distribución: '+str(np.std(Dist_interg)))
```

La diferencia entre el valor real de 63 y la media de 80.11 es de casi tres desviaciones estándar por debajo. Esto sugiere fuertemente que el comportamiento de la red real no se debe al azar y que es efectivamente homofílica. Para corroborarlo se analizará el p-valor en el ítem siguiente.

#### Ítem iii: Estimación de significancia

Lo que hacemos a continuación es calcular el p-valor, entendido como la probabilidad acumulada de tener, por azar un valor igual o menor que el valor obtenido en la realidad. El p-valor lo hemos calculado como la frecuencia de los valores iguales o menores que el valor real, dividido por la cantidad total de valores. Como inclusión final se puede observar el desplazamiento del valor real (línea roja), con relación al valor medio (máximo de la distribución).

```python
# Determinamos los valores que son menores al valor real.

p_v =[]
d = (Dist_interg)
for x in range(0,len(d)):
    if d[x]<=63:
        p_v.append(d[x])
        
# Hacemos la cuenta de lo que vale el p-valor

print('p-valor de la distribución: '+str(sum(p_v)/sum(d)))
```

Este p-valor es más pequeño que el usualmente aceptado, de lo cual se concluye que la red es homofílica, ya que la probabilidad de que la distribución real de enlaces sea resultado del azar es de alrededor del 0.5 %.

### Ítem b: Estimación del exponente

Ahora se estimará el exponente de la distribución a partir del gráfico de bineado logarítmico en la escala log-log.

```python
import igraph as ig
fit=ig.power_law_fit(Datos)
gamma=fit.alpha #Este es el exponente de la power law
Kmin=fit.xmin #Este el K_min
print(gamma)
print(Kmin)

b=np.logspace(0,3,20)
p=(gamma-1)/Kmin*(b/Kmin)**(-gamma)

plt.hist(Datos, bins=b, density=True, log=True) 
plt.xscale('log')
plt.plot(b,p)
plt.show()

print('Exponente de la distribución: '+str(gamma))
print('K mínimo: '+str(Kmin))
```

Dado que el eponente gamma es superior a 2, es posible concluir que el sitema presenta una distribución libre de escala.

```python
# Definimos la función para calcular la asortatividad de una red.

def asortatividad(archivo):

    H = nx.Graph()
    H.add_edges_from(archivo)
    B = nx.to_pandas_adjacency(H)
    BN = B.values
    f = H.degree # Asigna un grado distinto a cada uno de los nodos.
    grados_H_d = [gra for (nom,gra) in f] # Se queda con el valor del grado para cada uno de los nodos. Genera una tira de grados.
    grados_H = np.asarray(grados_H_d) # Es la forma array de todos los grados que hemos calculado en las funciones anteriores.

    b = np.dot(BN,grados_H)

    cuent= np.dot(BN,np.asarray([1]*len(BN)))
    c = b/cuent

    d = [0]*max(grados_H)
    e = range(max(grados_H))

    for i in e:
        sumo = 0
        cont = 0
        for k in range(len(grados_H)):
            if grados_H[k] == i:
                sumo = sumo+c[k]
                cont = cont+1
        if cont != 0 and sumo != 0:
            d[i] = sumo/cont
        else:
            d[i] = 1
    f = np.log(d[1:11])
    g = np.log(e[1:11])       
    fitfunc = lambda p, g: p[0]*g+p[1]# Funcion con la que queremos fitear la data
    p0 = [1, 1] # Parametros iniciales para fitear
    errfunc = lambda p, g, f: fitfunc(p, g) - f # Distancia a la función objetivo
    p1, success = optimize.leastsq(errfunc, p0[:], args=(g, f))
    plt.plot(g, f, "ro", g, fitfunc(p1, g), "k-", linewidth=3) # Plot of the data and 
    return p1[0]
```
