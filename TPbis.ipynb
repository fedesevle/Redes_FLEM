```python
import matplotlib.pylab as plt
import networkx as nx

def ldata(archive):
    f=open(archive)
    data=[]
    for line in f:
        line=line.strip()
        col=line.split()
        data.append(col)
    return data

a = ldata("yeast_AP-MS.txt")

G = nx.Graph()
G.add_edges_from(a)
nx.draw(G, node_size=5)
plt.show()

b = ldata("yeast_Y2H.txt")

H = nx.Graph()
H.add_edges_from(a)
nx.draw(H, node_size=5)
plt.show()


c = ldata("yeast_LIT.txt")

I = nx.Graph()
I.add_edges_from(a)
nx.draw(I, node_size=5)
plt.show()
```
