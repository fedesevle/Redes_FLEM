```python
import matplotlib.pylab as plt
import networkx as nx
import numpy as np
import pandas as pd

#G = nx.read_gml('as-22july06.gml')
H = nx.read_gml('netscience.gml')

#A = nx.to_pandas_adjacency(G)
B = nx.to_pandas_adjacency(H)
#AN=A.values
BN=B.values

# print(BN)

f=H.degree
grados_H_d = [gra for (nom,gra) in f] # [lo_que_quiero  for los_elementos in iterable]
grados_H = np.asarray(grados_H_d)

# print(grados_H)
# print(b)

b=np.dot(BN,grados_H)

c=b/grados_H

# c=[]

# for i,(gg,bb) in enumerate(zip(grados_H,b)):
#     if gg!=0:
#         c.append(bb/gg)

# for i in range(len(grados_H)):
#     if grados_H[i]!=0:
#         c.append(b[i]/grados_H[i])

d=[0]*max(grados_H)
e=range(max(grados_H))

for i in e:
    sumo=0
    cont=0
    for k in range(len(grados_H)):
        if grados_H[k]==i:
            sumo=sumo+c[k]
            cont=cont+1
    if cont!=0:
        d[i]=sumo/cont
            
print(d)
    
plt.plot(e,d)
plt.show()
```
